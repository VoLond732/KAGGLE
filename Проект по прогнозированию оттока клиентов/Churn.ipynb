```python
# Импорт необходимых библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Настройка визуализаций
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
%matplotlib inline

# Загрузим данные
df = pd.read_csv('churn.csv')
```

 Проведем первичный осмотр данных

```python
# Размер датафрейма
print(f"Размер датасета: {df.shape[0]} строк, {df.shape[1]} столбцов")

# Первые 5 строк
print("\n Первые 5 строк данных:")
display(df.head())

# Информация о типах данных и пропусках
print("\n Информация о данных:")
df.info()

# Проверка пропущенных значений
print("\n Проверка пропущенных значений:")
missing = df.isnull().sum()
print(missing[missing > 0])

# Базовая статистика для числовых колонок
print("\n Базовая статистика (числовые колонки):")
display(df.describe())

# Проверка уникальных значений в каждой колонке
print("\n Количество уникальных значений в каждой колонке:")
for column in df.columns:
    unique_count = df[column].nunique()
    print(f"{column}: {unique_count} уникальных значений")
    if unique_count < 10:
        print(f"  Значения: {df[column].unique()}")
```

Целевая переменная - "Exited", посмотрим на нее поближе:

```python
# Распределение
exited_counts = df['Exited'].value_counts()
exited_percent = df['Exited'].value_counts(normalize=True) * 100

print(f"Распределение:")
print(f"0 (Не ушел): {exited_counts[0]} клиентов ({exited_percent[0]:.1f}%)")
print(f"1 (Ушел): {exited_counts[1]} клиентов ({exited_percent[1]:.1f}%)")

# Дисбаланс
imbalance_ratio = exited_counts[0] / exited_counts[1]
print(f"\n Дисбаланс классов (отношение оставшихся клиентов к ушедшим): {imbalance_ratio:.2f}")

# Визуализация
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Круговая диаграмма
ax1.pie(exited_counts.values, labels=['Не ушел (0)', 'Ушел (1)'], 
        autopct='%1.1f%%', colors=['lightblue', 'salmon'], explode=(0, 0.1))
ax1.set_title('Распределение оттока клиентов')

# Столбчатая диаграмма
sns.countplot(data=df, x='Exited', ax=ax2, palette='Set2')
ax2.set_title('Отток клиентов (количество)')
ax2.set_xticklabels(['Не ушел (0)', 'Ушел (1)'])
ax2.set_ylabel('Количество клиентов')
ax2.set_xlabel('')
plt.tight_layout()
plt.show()
```

Очистим данные и приведем категориальные признаки к бинарному виду

```python
df_processed = df.copy()

# Удаляем нефункциональные столбцы
cols_to_drop = ['RowNumber', 'CustomerId', 'Surname']
print(f"Удаляемые столбцы: {cols_to_drop}")
print("Причина: RowNumber и CustomerId - идентификаторы, Surname - фамилия, данные характеристики не влияют на отток")

df_processed = df_processed.drop(columns=cols_to_drop)
print(f" Удалено {len(cols_to_drop)} столбца")
print(f" Новая размерность данных: {df_processed.shape}")

# Проверяем типы данных
print("\n ТИПЫ ДАННЫХ ПОСЛЕ ОЧИСТКИ:")
print(df_processed.dtypes)

# Разделяем признаки на категориальные и числовые
categorical_cols = df_processed.select_dtypes(include=['object']).columns.tolist()
numeric_cols = df_processed.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Убираем целевую переменную из числовых признаков
numeric_cols.remove('Exited')

print(f"\n Категориальные признаки ({len(categorical_cols)}): {categorical_cols}")
print(f" Числовые признаки ({len(numeric_cols)}): {numeric_cols}")

# Проверяем уникальные значения в категориальных признаках
print("\n УНИКАЛЬНЫЕ ЗНАЧЕНИЯ В КАТЕГОРИАЛЬНЫХ ПРИЗНАКАХ:")
for col in categorical_cols:
    unique_vals = df_processed[col].unique()
    value_counts = df_processed[col].value_counts()
    print(f"\n{col}:")
    print(f"  Уникальных значений: {len(unique_vals)}")
    print(f"  Значения: {unique_vals}")
    print(f"  Распределение:")
    for val, count in value_counts.items():
        print(f"    {val}: {count} ({count/len(df_processed)*100:.1f}%)")

#Переведем категориальные признаки в числовые
geo_dummies = pd.get_dummies(df_processed['Geography'], prefix='Geo', drop_first=False, dtype='int64')
df_processed = pd.concat([df_processed, geo_dummies], axis=1)
df_processed = df_processed.drop('Geography', axis=1)

gender_mapping = {'Female': 0, 'Male': 1}
df_processed['Gender'] = df_processed['Gender'].map(gender_mapping)

# Сохраняем очищенные данные
df_processed.to_csv('processed_churn.csv', index=False)
print(f"\n Размер очищенного датасета: {df_processed.shape}")
```

Проведем статистический анализ признаков

```python
from scipy import stats
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Выбираем числовые признаки (исключаем целевую переменную и бинарные)
continuous_features = ['CreditScore', 'Age', 'Balance', 'EstimatedSalary']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for idx, feature in enumerate(continuous_features):
    ax = axes[idx]
    
    # Гистограмма с KDE
    sns.histplot(data=df_processed, x=feature, ax=ax, kde=True, 
                color='steelblue', bins=30, edgecolor='black')
    
    mean_val = df_processed[feature].mean()
    median_val = df_processed[feature].median()
    
    ax.axvline(mean_val, color='red', linestyle='--', linewidth=2, 
              label=f'Среднее: {mean_val:.1f}')
    ax.axvline(median_val, color='green', linestyle='--', linewidth=2,
              label=f'Медиана: {median_val:.1f}')
    
    ax.set_title(f'Распределение {feature}', fontsize=14, fontweight='bold')
    ax.set_xlabel(feature, fontsize=12)
    ax.set_ylabel('Частота', fontsize=12)
    ax.legend()
    ax.grid(True, alpha=0.3)

plt.suptitle('Распределение непрерывных признаков', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()


print(" Статистика непрерывных признаков:")
for feature in continuous_features:
    skewness = df_processed[feature].skew()
    kurtosis = df_processed[feature].kurtosis()
    print(f"\n{feature}:")
    print(f"  Асимметрия (skewness): {skewness:.3f} ({'симметрично' if -0.5 < skewness < 0.5 else 'скошено вправо' if skewness > 0.5 else 'скошено влево'})")
    print(f"  Эксцесс (kurtosis): {kurtosis:.3f} ({'нормальное' if -0.5 < kurtosis < 0.5 else 'тяжелые хвосты' if kurtosis > 0.5 else 'легкие хвосты'})")
```

Рассмотрим влияние признаков на отток клиентов

```python

fig, axes = plt.subplots(3, 4, figsize=(18, 12))

# Список всех признаков кроме целевой переменной
features_to_analyze = [col for col in df_processed.columns if col != 'Exited']

# Для каждого признака строим график
for idx, feature in enumerate(features_to_analyze):
    row = idx // 4
    col = idx % 4
    ax = axes[row, col]
    
    # Определяем тип графика в зависимости от признака
    unique_vals = df_processed[feature].nunique()
    
    if unique_vals <= 5:  # Категориальные/бинарные признаки
        # Bar plot со средним оттоком
        churn_by_feature = df_processed.groupby(feature)['Exited'].mean() * 100
        bars = ax.bar(churn_by_feature.index.astype(str), churn_by_feature.values,
                     color=['lightblue' if x < 20 else 'salmon' for x in churn_by_feature.values])
        
        
        for bar, val in zip(bars, churn_by_feature.values):
            height = bar.get_height()
            ax.text(bar.get_x() + bar.get_width()/2., height + 0.5,
                   f'{val:.1f}%', ha='center', va='bottom', fontsize=9)
        
        ax.set_ylim(0, max(churn_by_feature.values) * 1.2)
        
    else:  # Непрерывные признаки
        sns.boxplot(data=df_processed, x='Exited', y=feature, ax=ax, palette='Set2')
        ax.set_xticklabels(['Не ушел', 'Ушел'])
    
    ax.set_title(feature, fontsize=12, fontweight='bold')
    ax.set_xlabel('' if unique_vals > 5 else feature)
    ax.set_ylabel('Отток (%)' if unique_vals <= 5 else feature)
    ax.grid(True, alpha=0.3)
    ax.tick_params(axis='x', rotation=45 if unique_vals <= 5 else 0)

for idx in range(len(features_to_analyze), 12):
    row = idx // 4
    col = idx % 4
    axes[row, col].axis('off')

plt.suptitle('Влияние признаков на отток клиентов', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

Проанализируем взаимосвязи между признаками, их корреляцию с оттоком клиентов

```python
# Создаем матрицу корреляций
correlation_matrix = df_processed.corr()

# Визуализируем полную матрицу корреляций
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Матрица корреляций всех признаков', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Корреляция с целевой переменной
correlation_with_target = correlation_matrix['Exited'].sort_values(ascending=False)

print(" КОРРЕЛЯЦИЯ ПРИЗНАКОВ С ЦЕЛЕВОЙ ПЕРЕМЕННОЙ (Exited):")
print("-" * 60)

for feature, corr in correlation_with_target.items():
    if feature == 'Exited':
        continue
    
    # Интерпретация корреляции
    if abs(corr) >= 0.3:
        strength = "Значительная"
    elif abs(corr) >= 0.2:
        strength = "Умеренная"
    elif abs(corr) >= 0.1:
        strength = "Слабая"
    else:
        strength = "Незначительная"
    
    direction = "увеличивает" if corr > 0 else "уменьшает"
    
    print(f"{feature:20} : {corr:+.4f} | {strength} | {direction} вероятность оттока")

# Визуализация корреляций
top_corr = correlation_with_target.drop('Exited').abs().sort_values(ascending=False).head(12)

plt.figure(figsize=(10, 6))
top_corr.sort_values().plot(kind='barh', label='_nolegend_', color=['salmon' if x > 0 else 'lightblue' 
                                               for x in correlation_with_target.loc[top_corr.index]])
plt.title('Градация признаков по абсолютной корреляции с оттоком\n (синий цвет - корреляция с оттоком положительная, красный - отрицательная)', fontsize=14, fontweight='bold')
plt.xlabel('Абсолютная корреляция', fontsize=12)
plt.ylabel('Признак', fontsize=12)
plt.axvline(x=0.3, color='red', linestyle='--', alpha=0.7, label='Значительная корреляция (0.3)')
plt.axvline(x=0.2, color='orange', linestyle='--', alpha=0.7, label='Умеренная корреляция (0.2)')
plt.axvline(x=0.1, color='green', linestyle='--', alpha=0.7, label='Слабая корреляция (0.1)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Проверим значимость различий между ушедшими и оставшимися клиентами

```python
print(" СТАТИСТИЧЕСКИЕ ТЕСТЫ: Различия между ушедшими и оставшимися клиентами")
print("=" * 80)

# Разделяем данные
churned = df_processed[df_processed['Exited'] == 1]
stayed = df_processed[df_processed['Exited'] == 0]

for feature in ['Age', 'CreditScore', 'Balance', 'EstimatedSalary']:
    # t-тест для сравнения средних
    t_stat, p_value = stats.ttest_ind(churned[feature], stayed[feature])
    
    print(f"\n Признак: {feature}")
    print(f"   Ушедшие:   Среднее = {churned[feature].mean():.2f}, Std = {churned[feature].std():.2f}")
    print(f"   Оставшиеся: Среднее = {stayed[feature].mean():.2f}, Std = {stayed[feature].std():.2f}")
    print(f"   Разница: {churned[feature].mean() - stayed[feature].mean():.2f}")
    
    if p_value < 0.001:
        significance = "*** Высоко значимо"
    elif p_value < 0.01:
        significance = "** Значимо"
    elif p_value < 0.05:
        significance = "* Умеренно значимо"
    else:
        significance = "Не значимо"
    
    print(f"   t-статистика: {t_stat:.4f}, p-value: {p_value:.6f} {significance}")

# Хи-квадрат тест для категориальных признаков
print("\n Хи-квадрат тест для категориальных признаков:")
for feature in ['Gender', 'Geo_Germany', 'Geo_Spain', 'HasCrCard', 'IsActiveMember']:
    contingency_table = pd.crosstab(df_processed[feature], df_processed['Exited'])
    chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)
    
    print(f"\n   Признак: {feature}")
    print(f"   Хи-квадрат: {chi2:.4f}, p-value: {p_value:.6f}")
    print(f"   Таблица сопряженности:")
    print(f"   {contingency_table}")
```

Подведем промежуточные итоги:

```python
print(" КЛЮЧЕВЫЕ ВЫВОДЫ ИЗ ИССЛЕДОВАТЕЛЬСКОГО АНАЛИЗА ДАННЫХ")
print("=" * 80)

# Собираем ключевые инсайты
insights = []

#  Возраст
age_churn_rate = df_processed.groupby(df_processed['Age'] // 10 * 10)['Exited'].mean()
max_age_group = age_churn_rate.idxmax()
max_age_rate = age_churn_rate.max() * 100
insights.append(f"1. **Возраст**: Клиенты в возрасте {max_age_group}+ лет имеют максимальный отток ({max_age_rate:.1f}%)")

# География
geo_cols = ['Geo_Germany', 'Geo_Spain']
for col in geo_cols:
    if col in df_processed.columns:
        churn_rate = df_processed[df_processed[col] == 1]['Exited'].mean() * 100
        country = 'Германия' if 'Germany' in col else 'Испания'
        insights.append(f"2. **География**: Клиенты из {country} имеют отток {churn_rate:.1f}%")

# Активность
active_churn = df_processed.groupby('IsActiveMember')['Exited'].mean() * 100
insights.append(f"3. **Активность**: Неактивные клиенты уходят в {active_churn[0]/active_churn[1]:.1f} раза чаще")

# Баланс
balance_corr = df_processed['Balance'].corr(df_processed['Exited'])
insights.append(f"4. **Баланс**: Корреляция баланса с оттоком: {balance_corr:.3f}")

# Количество продуктов
products_churn = df_processed.groupby('NumOfProducts')['Exited'].mean() * 100
max_products = products_churn.idxmax()
max_products_rate = products_churn.max()
insights.append(f"5. **Продукты**: Клиенты с {max_products} продуктами имеют отток {max_products_rate:.1f}%")

# Выводим все инсайты
print("\n".join(insights))

# Сохраняем ключевые выводы в файл
with open('eda_insights.txt', 'w', encoding='utf-8') as f:
    f.write("КЛЮЧЕВЫЕ ВЫВОДЫ ИЗ EDA\n")
    f.write("=" * 50 + "\n\n")
    f.write("\n".join(insights))

print(f"Выводы сохранены")
```
